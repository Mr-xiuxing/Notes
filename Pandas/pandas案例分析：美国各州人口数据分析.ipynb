首先导入文件，并查看数据样本

```python
import pandas as pd
```

```python
pop = pd.read_csv("./state-population.csv")
pop.shape
```

```python
pop.head()
```

```python
areas = pd.read_csv("./state-areas.csv")
areas.head()
```

```python
abbr = pd.read_csv("./state-abbrevs.csv")
abbr.head()
```

合并pop与abbrevs两个DataFrame，分别依据state/region列和abbreviation列来合并。

为了保留所有信息，使用外合并。

```python
# 一对多的合并
pop2 = pop.merge(abbr,left_on="state/region",right_on="abbreviation",how="left")
pop2.head()
```

```python
pop2.shape
```

去除abbreviation的那一列（axis=1）

```python
pop2.drop("abbreviation",axis=1,inplace=True)
pop2.head()
```

查看存在缺失数据的列。

使用.isnull().any()，只有某一列存在一个缺失数据，就会显示True。

pop2.isnull().any(axis = 0)

```python
pop2.isnull().any(axis=0)
```

查看缺失数据

根据数据是否缺失情况显示数据，如果缺失为True，那么显示

```python
cond = pop2.isnull().any(axis=1)
pop2[cond]
```

找到有哪些state/region使得state的值为NaN，使用unique()查看非重复值

```python
cond_state = pop2["state"].isnull()
cond_state
```

```python
pop2[cond_state]["state/region"].unique() #unique:去重函数  PR和USA 这两个州的全称是空值
```

为找到的这些state/region的state项补上正确的值，从而去除掉state这一列的所有NaN！

记住这样清除缺失数据NaN的方法！

```python
cond2 = pop2["state/region"] == "PR"
pop2[cond2].shape
```

```python
cond2.sum()
```

```python
pop2["state"][cond2] = "Puerto Rico"
```

```python
cond3 = pop2["state/region"] == "USA"
cond3.sum()
```

```python
pop2["state"][cond3] = "The United State"
```

```python
pop2.isnull().any()
```

合并各州面积数据areas，使用左合并。

```python
pop2.head()
```

```python
areas.head()
```

```python
pop3 = pop2.merge(areas, how="left")
pop3.shape
```

继续寻找存在缺失数据的列

```python
pop3.isnull().any()
```

我们会发现area(sq.mi)这一列有缺失数据，为了找出是哪一行，我们需要找出是哪个state没有数据

```python
cond4 = pop3["area (sq. mi)"].isnull()
cond4
```

```python
usa_areas = areas["area (sq. mi)"].sum()
usa_areas
```

```python
pop3["area (sq. mi)"][cond4] = usa_areas
pop3.isnull().any()
```

去除含有缺失数据的行

```python
pop3.dropna(inplace=True)
```

查看数据是否缺失

```python
pop3.isnull().any()
```

```python
pop3.shape
```

找出2010年的全民人口数据,df.query(查询语句)

```python
pop3.head()
```

```python
pop_2010 = pop3.query("year == 2010 & ages == 'total'")
pop_2010
```

对查询结果进行处理，以state列作为新的行索引:set_index

```python
# set_index 很重要的， 通过设置时间为行索引， 可以进行数据可视化
pop_2010.set_index("state",inplace=True)
pop_2010
```

```python
pop_2010.loc["Puerto Rico"]
```

计算人口密度。注意是Series/Series，其结果还是一个Series。

```python
pop_dense = pop_2010["population"]/pop_2010["area (sq. mi)"]
pop_dense
```

排序，并找出人口密度最高的五个州sort_values()

```python
pop_dense.sort_values(inplace=True)
pop_dense.head()
```
