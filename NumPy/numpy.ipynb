```python
"""
4. 级联
np.concatenate() 级联需要注意的点：
级联的参数是列表：一定要加中括号或小括号
维度必须相同
形状相符
【重点】级联的方向默认是shape这个tuple的第一个值所代表的维度方向
可通过axis参数改变级联的方向
"""
```

```python
# 级联:就是两个ndarray的数据 合到一起
nd1 = np.array([1,2,3])
nd2 = np.array([1,2,3])
np.concatenate([nd1,nd2])
```

```python
# 二维数据进行级联  轴没有匹配 axis = 0   在行上面进行级联 增加行数
nd1 = np.random.randint(0,10,size = (2,3))
nd2 = np.random.randint(50,100,size = (4,3))
np.concatenate([nd1, nd2])                  
```

```python
# 匹配轴 axis = 1   在行上面进行级联 增加行数
nd1 = np.random.randint(0,10,size = (4,6))
nd2 = np.random.randint(50,100,size = (4,3))
np.concatenate([nd1, nd2], axis = 1)
```

```python
#行上面进行级联的时候（axis = 0）, 列是必须相同的
#列上面进行级联的时候（axis = 1）, 行是必须相同的
```

```python
np.hstack与np.vstack
水平级联与垂直级联,处理自己，进行维度的变更
```

```python
# 六、ndarray的排序
```

```python
nd = np.random.randint(0,100,size=10)
nd
```

```python
def sort_nd(nd):
    for i in range(nd.size):
        for j in range(i, nd.size):
            if nd[i] > nd[j]:
                nd[i], nd[j] = nd[j], nd[i]
    return nd
```

```python
sort_nd(nd)
```

```python
nd
```

```python
nd2 = np.random.randint(0,100,size=10)
nd2
```

```python
def sort_nd2(nd):
    for i in range(nd2.size):
        index_min = np.argmin(nd2[i:]) + i
        nd2[i],nd2[index_min] = nd2[index_min],nd2[i]
    return nd2
```

```python
sort_nd2(nd2)
```

```python
# 部分排序
"""
np.partition(a,k)

有的时候我们不是对全部数据感兴趣，我们可能只对最小或最大的一部分感兴趣。

当k为正时，我们想要得到最小的k个数
当k为负时，我们想要得到最大的k个数
"""
```

```python
nd = np.random.randint(0,10000, size = 500)
nd
```

```python
#就是把前kth 个数字 排在前面
np.partition(nd, kth = 10)
```

```python
np.partition(nd,kth=-10)[-10:]
```
